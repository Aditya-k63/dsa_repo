```python
###   HashSet

nums=[100,4,200,1,3,2]

num_set=set(nums)
longest=0

for n in num_set:
    if n-1 not in num_set:
        length=1

        while n+ length in num_set:
            length+=1
        longest=max(longest,length)
print(longest)
```

```python
## finding te max value of the arrr

def find_max(arr):
    mx=arr[0]

    for i in arr:
        if i > mx:
            mx=i
    return mx

arr=[4,9,2,15,7]
print(find_max(arr))
```

```python
##count the even number 

def count_even(arr):
    eve=0
    for i in arr:
        if i %2==0:
            eve+=1
    return eve
arr = [3, 8, 5, 12, 7, 10]
print(count_even(arr))
```

```python
##  check if array is sorted 

def sorted_check(arr):
    for i in range (len(arr)-1):
        if arr[i]>arr[i+1]:
            return False
    return True

arr = [1, 3, 2, 4, 5]
print(sorted_check(arr))
```

```python
##  second largets number

def second_largest (arr):
    if len(arr)<2:
        return None
    
    largest=float("-inf")
    second=float("-inf")

    for i in arr:
        if i > largest:
            second=largest
            largest=i

        elif largest > i > second:
            second = i
    if second==float("-inf"):
        return None
    return second

arr = [10, 5, 20, 8, 15]
print(second_largest(arr))
```

```python
##  counting frequency of each elemrnt 

def count_frequency(arr):
    freq={}

    for d in arr:
        freq[d]=freq.get(d,0)+1
    return freq
arr = [4, 5, 4, 3, 5, 5]
print(count_frequency(arr))
```

```python
## check for duplicate 

def contains_duplicate(arr):
    seen=set()

    for i in arr:
        if i in seen:
            return True
        seen.add(i)
    return False

arr = [1, 3, 5, 3, 7]
print(contains_duplicate(arr))
```

```python
## Q7 — Find the Most Frequent Element

def most_frequent(arr):
    freq={}


    for i in arr:
        freq[i]=freq.get(i,0)+1
    return max(freq,key=freq.get)

arr = [1, 2, 2, 3, 3, 3, 4]
print(most_frequent(arr))
```

```python
##  valid anagram 

def valid_anagram(s,t):
    if len(s)!=len(t):
        return False
    freq={}

    for c in s:
        freq[c]=freq.get(c,0)+1
    
    for c in t :
        if c not in freq:
            return False
        freq[c]-=1

        if freq[c]<0:
            return False
    return True
  
s = "listen"
t = "silent"

print(valid_anagram(s,t))

```

```python
## two sum 

def two_sum(nums,target):
    seen={}

    for i, num in enumerate(nums):
        need=target-num

        if need in seen:
            return[seen[need],i]
        
        seen[num]=i
    return []
    
nums = [2, 7, 11, 15]
target = 9

print(two_sum(nums,target))
```

```python
##  Q10 — First Non-Repeating Character

def non_repeating(s):
    freq={}

    for c in s:
        freq[c]=freq.get(c,0)+1
    for c in s :
        if freq[c]==1:
            return c 
    return None

s = "aabbcde"

print(non_repeating(s))
```
